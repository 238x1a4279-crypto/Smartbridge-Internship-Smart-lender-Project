```python
import pandas as pd

df = pd.read_csv('/content/loan_prediction.csv')  # Replace with the actual filename if different
df.head()
```

**Import Libraries**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')
```

**Bivariate Analysis**

Bivariate analysis examines the relationship between two variables.

**Gender vs Married**

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Gender', hue='Married', data=df)
plt.title('Gender vs Married')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**Education vs Self_Employed**

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Education', hue='Self_Employed', data=df)
plt.title('Education vs Self Employed')
plt.xlabel('Education')
plt.ylabel('Count')
plt.show()
```

**Property Area vs Loan Amount Term**

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Property_Area', y='Loan_Amount_Term', data=df)
plt.title('Property Area vs Loan Amount Term')
plt.xlabel('Property Area')
plt.ylabel('Loan Amount Term')
plt.show()
```

**Gender vs Loan Status**

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Gender', hue='Loan_Status', data=df)
plt.title('Gender vs Loan Status')
plt.show()
```

**Observations**



*   Bivariate analysis was performed to study relationships between pairs of variables.
*   The Gender vs Married plot shows the distribution of marital status across genders.

*   The Education vs Self_Employed plot highlights the relationship between education level and employment type.
*   The Property Area vs Loan Amount Term plot compares loan repayment terms across different property areas.

*   Additional comparisons such as Gender vs Loan Status and Credit History vs Loan Status help identify features that may influence loan approval.
