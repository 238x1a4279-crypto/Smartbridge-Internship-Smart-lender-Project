```python
import pandas as pd

df = pd.read_csv('/content/loan_prediction.csv')  # Replace with the actual filename if different
df.head()
```

**Import Libraries**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')
```

**Check the column names**

```python
print(df.columns)
```

**Distribution Plot (Continuous Variable)**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['ApplicantIncome'], kde=True)
plt.title("Distribution of Applicant Income")
plt.xlabel("Applicant Income")
plt.ylabel("Frequency")
plt.show()
```

**Count Plot for Gender**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

**Count Plot for Education**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Education', data=df)
plt.title("Education Distribution")
plt.show()
```

**Count Plot for Credit History**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Credit_History', data=df)
plt.title("Credit History")
plt.show()
```

**Multiple Count Plots Using Subplots**

```python
categorical_features = ['Gender','Married','Education','Self_Employed','Property_Area']

plt.figure(figsize=(15,10))

for i, feature in enumerate(categorical_features):
    plt.subplot(2,3,i+1)
    sns.countplot(x=feature, data=df)

plt.tight_layout()
plt.show()
```

**Observations**

After generating the plots, include observations such as:



*   Applicant income distribution is skewed, indicating that most applicants have lower to moderate incomes while a smaller number have higher incomes.


*   Credit history is a binary feature with values 0 and 1.
*   Gender and Education are categorical variables with two categories each.

*   The count plots show the frequency of each category.
*   Univariate analysis helps understand the distribution of individual features before data preprocessing and model building.
